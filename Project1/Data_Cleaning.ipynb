```python
columns = ["case_id", "t_state", "p_name", "p_year", "p_tnum", "p_cap", "xlong", "ylat"]
data = filedf[columns]
data = data.dropna(how="any")
```

```python
data.drop_duplicates(subset = "p_name", inplace = True)
data
filtered_df = data.loc[data["p_year"] > 2001.0]
# fltrdSorted_df = filtered_df.sort_values("p_year")
```

```python
# #group by year and sum of p_cap through those years
groupyear = filtered_df.groupby(["p_year"]).mean()["p_cap"]
groupyeardf = pd.DataFrame(groupyear)
x= groupyeardf.reset_index()
#graph year(x) cap (y)
x.head()
x.rolling(2, on = "p_cap").sum()
x.to_csv("Output/windcap.csv", index = True, header = True)
```
